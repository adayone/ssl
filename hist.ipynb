```python
import tushare as ts
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import matplotlib
matplotlib.style.use('ggplot')
```

```python
id = '000007'
df = ts.get_h_data(id, autype='qfq', start='2013-06-10')
```

```python
close = pd.DataFrame(df.query('date > "2015-01-01"')['close'])
```

```python
m1m = pd.rolling_mean(close, window=30, min_periods=1, center=True)
m2m = pd.rolling_mean(close, window=60, min_periods=1, center=True)
m4m = pd.rolling_mean(close, window=120, min_periods=1, center=True)
wdf = close.join(m1m, how='left', rsuffix='_m1') \
.join(m2m, how='left',  rsuffix='_m2') \
.join(m4m, how='left', rsuffix='_m4') 
```

```python
print wdf.plot(title=id)
```
