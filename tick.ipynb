```python
import tushare as ts
import datetime as dt
%matplotlib inline

import pandas as pd
import matplotlib
matplotlib.style.use('ggplot')

```

```python
# get all time ticks
def get_ticks(id, end, delta):
    if isinstance(end, str):
        end = dt.datetime.strptime(end_str, '%Y-%m-%d')
    date_list = [end - dt.timedelta(days=x) for x in range(0, 100)]
    ticks = None
    for date in date_list:
        date_str = dt.datetime.strftime(date, '%Y-%m-%d')
        try:
            tick = ts.get_tick_data(id, date_str)
        except:
            continue
        if tick is None:
            continue
        tick['date'] = date_str
        if ticks is None:
            ticks = tick
        else:
            ticks = ticks.append(tick)
        ticks['dt'] = ticks['date'] + ' '+ ticks['time']
        ticks['id'] = id

    return ticks
```

```python
def cmp_top(ticks):
    if ticks is None:
        print 'no'
        return None
    try:
        today = ts.get_today_ticks(id).sort('amount', ascending=False)
    except:
        return None
    sticks = ticks.sort('amount', ascending=False).head(20)
    sticks = sticks.reset_index()
    today = today.reset_index()
    top = sticks.ix[0]
    top_today = today.ix[0]

    return [(round((top.price - top_today.price)/top.price, 3)), round(top.price, 3), round(top_today.price, 3), top.amount, top_today.amount]
```

```python
def tick_plot(ticks, end_str, delta=100): 
    # init
    ticks = ticks.sort('dt', ascending=True)

    fig = ticks.plot(x='dt', y=['price', 'amount'],  secondary_y='amount',  legend=True, title=id, rot=30, figsize=(16,6)) 
    return fig.get_figure()
```

```python
end_str = '2015-07-20'
stocks = ts.get_stock_basics()
for id in ('000070', '000005', '002747', '300104', '300017', '002594','000762', '600773', '000626', '600804'):
    ticks = get_ticks(id, end_str, 100)
    name = stocks.ix[id]['name']
    ticks['name'] = name
    fig = tick_plot(ticks, end_str)
    fig.savefig('image/%s.png'%name)
    sticks = ticks.sort('amount', ascending=False).head(5)
    sticks.to_csv('data/%s.csv'%name,  index=None)

```
