## LR
####sigmoid函数

$\frac{1}{1 + e^{-z}}$

#### 输入z
$z=w_0x_0 + w_1x_0 + ... + w_nx_n$

### 梯度上升

伪码：

1. 计算整个数据集的梯度
2. 使用 $\alpha \times gradient$更新回归系数的向量
3. 返回回归系数

```python
import numpy as np
```

```python
def loadDataSet():
    data = []
    label = []
    fr = open('testSet.txt')
    for line in fr.readlines():
        items = line.strip().split()
        data.append([1.0, float(items[0],), float(items[1])])
        label.append(int(items[2]))
    fr.close()
    return data, label
                        
```

```python
def sigmoid(x):
    return 1.0/(1+np.exp(-x))
```

```python
def gradAscent(data, label):
    data_mat = np.mat(data)
    label_mat = np.mat(label).transpose()
    m, n = np.shape(data_mat)
    alpha = 0.001
    max_iter = 200
    weights = np.ones((n, 1))
    for k in range(max_iter):
        h = sigmoid(data_mat * weights)
        error = label_mat - h
        weights = weights + alpha * data_mat.transpose() * error
    return weights
                 
                            
```

```python
data, label = loadDataSet()
```

```python
weight = gradAscent(data, label)
```

### 决策边界

边界方程式：
$ 0 = w_0x_0 + w_1x_1 + w_2x_2$

```python
import  pylab as plt
def bestPlot(weights):
    data, label = loadDataSet()
    data_arr = np.array(data)
    n = np.shape(data_arr)[0]
    xcord1 = []
    ycord1 = []
    xcord2 = []
    ycord2 = []
    
    for i in range(n):
        if int(label[i]) == 1:
            xcord1.append(data_arr[i, 1])
            ycord1.append(data_arr[i, 2])
        else:
            xcord2.append(data_arr[i, 1])
            ycord2.append(data_arr[i, 2])
    fig = plt.figure()
    ax = fig.add_subplot(111)
    # 两类的数据点
    ax.scatter(xcord1, ycord1, s=30, c='red', marker='s')
    ax.scatter(xcord2, ycord2, s=30, c='green')
    x = np.arange(-3.0, 3.0, 0.1)
    # 决策边界方程
    y = (-weights[0] - weights[1] * x)/weights[2]
    ax.plot(x, y)
    plt.xlabel('X1')
    plt.ylabel('X2')
    plt.show()
        
                          
                          
               
```

### SGD
梯度上升需要遍历整个数据集， 随机梯度上升则每次只利用一个样本点来更新回归系数。

```python
def randomGradAscent(data, label):
    m, n = np.shape(data)
    alpha = 0.01
    weights = np.ones(n)
    for i in range(m):
        h = sigmoid(sum(data[i] * weights))
        error = label[i] - h
        weights = weights + alpha * error * data[i]
    return weights
```

```python
data, label = loadDataSet()
```

```python
weight = randomGradAscent(np.array(data), label)
```

```python
bestPlot(weight)
```

差不多分错了1/3的样本， 但是梯度上升是在数据上迭代了200次得到的结果。

### SGD

实际上的onlinelearning是没有迭代次数的概念的， 但是我们可以利用迭代次数来模拟更多的输入数据。


```python
import random
def randomGradAscentAdv(data, label, iter_num = 2):
    m, n = np.shape(data)
    mRange = range(m)
    weight_init = np.ones(n)
    for j in range(iter_num):
        for i in range(m):
            alpha = 4/(1.0 + j + i) + 0.01
            ix = int(random.uniform(0, m))
            h = sigmoid(sum(data[ix] * weight_init))
            error = label[ix] - h
            weight_init = weight_init + alpha * error * data[ix]
    return weight_init
```

```python
data, label = loadDataSet()
weight = randomGradAscentAdv(np.array(data), label)
bestPlot(weight)
```

### learning rate - 跌代次数的开方

learning rate = $\sqrt{i}$

```python
import random
import math
def randomGradAscentAdvSqrt(data, label, iter_num = 2):
    m, n = np.shape(data)
    mRange = range(m)
    weights = np.ones(n)
    for j in range(iter_num):
        for i in range(m):
            alpha = 1/math.sqrt(i + j + 1)
            ix = int(random.uniform(0, m))
            h = sigmoid(sum(data[ix] * weights))
            error = label[ix] - h
            weights = weights + alpha * error * data[ix]
    return weights

```

```python
data, label = loadDataSet()
weight = randomGradAscentAdvSqrt(np.array(data), label)
bestPlot(weight)
```

## SGD - per coord learning rate

来自论文《Ad Click Predicition, a View from the Trenches》

从结果上看， 小数据情况下改学习速率没有多大的优势。

```python
import math
def randomGradAscentAdvPerCoord(data, label, iter_num = 100):
    m, n = np.shape(data)
    alpha = 0.01
    belta = 1
    mRange = range(m)
    weights = np.ones(n)
    sh = [0.0, 0.0, 0.0]
    for j in range(iter_num):
        for i in range(m):
            ix = int(random.uniform(0, m))
            h = sigmoid(sum(data[ix] * weights))
            error = label[ix] - h
            d = error * data[ix]
            sh += np.power(d, 2)
            nalpha = alpha / ( belta + np.sqrt(sh)) 
            weights = weights + nalpha * error *  data[ix]
    return weights
```

```python
data, label = loadDataSet()
weight = randomGradAscentAdvPerCoord(np.array(data), label)
bestPlot(weight)
```

## 分类

LR的回归还是比较简单的， 根据回归系数计算每个特征的输出， 然后求和， 最后输入到sigmoid函数。

大于0.5就是1， 否则就是0。

```python
def classify(data, weight):
    prob = sigmoid(sum(data * weight))
    if prob > 0.5: return 1.0
    else: return 0.0
```

```python
def load_cross_data():
    trainset = open('lrTrain.txt')
    train_feature = []
    train_label = []
    for line in trainset.readlines():
        items = line.strip().split("\t")
        features = []
        # just get feature
        for i in range(21):
            features.append(float(items[i]))
        train_feature.append(features)
        train_label.append(float(items[21]))
    trainset.close()
    return train_feature, train_label
```

```python
def cross_validation_test(iter_num = 200):
    train_feature, train_label = load_cross_data()
    weights = randomGradAscentAdvIter(np.array(train_feature), train_label, iter_num)
    return errRate(weights)
   
```

```python
def errRate(weights):
    testset = open('lrTest.txt')
    train_label = []
    error_count = 0
    num_test_vec = 0.0    
    for line in testset.readlines():
        num_test_vec += 1.0
        items = line.strip().split('\t')
        features = []
        for i in range(21):
            features.append(float(items[i]))
        if int(classify(np.array(features), weights)) != int(items[21]):
            error_count += 1
        err_rate = (float(error_count)/num_test_vec)
    testset.close()
    return err_rate
```

```python
def multiTest(weights, iter_num=200):
    errSum = 0.0
    for k in range(iter_num):
        errSum += errRate(weights)
    return errSum/float(iter_num)
```

```python
import random
import math
def randomGradAscentAdvIter(data, label, iter_sample_num = 500):
    px = []
    py = []
    m, n = np.shape(data)
    mRange = range(m)
    weight_init = np.zeros(n)

    #alpha = 0.001 
    pre_err = 1.0
    for i in range(iter_sample_num):
        alpha = 4/(1.0 + i) + 0.001
        #alpha = 1/math.sqrt(i + 1)
        ix = int(random.uniform(0, m))
        h = sigmoid(sum(data[ix] * weight_init))
        error = label[ix] - h
        #if math.fabs(pre_err - error) < 0.001:
         #   break
        weight_init = weight_init + alpha * error * data[ix]
        px.append(i)
        py.append(multiTest(weight_init, 20))
        #pre_error = error
    return (weight_init, px, py)
```

```python
def randomGradAscentAdvPerCoordIter(data, label, iter_num = 100):
    px = []
    py = []
    m, n = np.shape(data)
    alpha = 0.01
    belta = 1
    mRange = range(m)
    weights = np.zeros(n)

    sh = np.zeros(n)

    for i in range(iter_num):

        ix = int(random.uniform(0, m))
        h = sigmoid(sum(data[ix] * weights))
        error = label[ix] - h
        d = error * data[ix]
        sh += np.power(d, 2)
        nalpha = alpha / ( belta + np.sqrt(sh)) #+ alpha
        weights = weights + nalpha * error *  data[ix]
        px.append(i)
        py.append(multiTest(weights, 20))
    return (weights, px, py)
```

```python
train_feature, train_label =load_cross_data()
weight_per, px_per, py_per = randomGradAscentAdvPerCoordIter(np.array(train_feature), train_label, 200)
weight_sgd, px_sgd, py_sgd = randomGradAscentAdvIter(np.array(train_feature), train_label, 200)
```

```python
def  normalization(x):
    # axis代表横轴和竖轴
    data_mean =  np.mean(x, axis=0)
    data_mean_adj = [1, data_mean[1], data_mean[2]]
    data_max = np.max(x, axis=0)
    data_nor = (x - data_mean)/data_max
    return data_nor
```
