```python

def get_feature(v):
    f_90 =  np.sum(v.visit_datetime >= np.datetime64("2014-04-15"))
    f_30 = np.sum(v.visit_datetime >= np.datetime64("2014-06-15"))
    f_7 = np.sum(v.visit_datetime >= np.datetime64("2014-07-08"))
    f_1 = np.sum(v.visit_datetime >= np.datetime64("2014-07-14"))
    v["f_90_"] = f_90
    v["f_30_"] = f_30
    v["f_7_"] = f_7
    v["f_1_"] = f_1
    v.drop("visit_datetime",axis=1, inplace=True)
    return v

grouped = feature.groupby(("user_id", "brand_id", "type")).apply(get_feature)
```

```python
def merge_type(v):
    sd = str(v.type)
    v["f_90_"+sd] = v.f_90_
    v["f_30_"+sd] = v.f_30_
    v["f_7_"+sd] = v.f_7_
    v["f_1_"+sd]=v.f_1_
    v.drop(["f_90_", "f_30_", "f_7_", "f_1_"],axis=1, inplace=True)
    return v
    
ub = grouped.groupby(("user_id", "brand_id")).apply(merge_type)
```
