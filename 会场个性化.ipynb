```python
import pandas
dataset = pandas.read_csv(r"/Users/haoyuanhu/Documents/时尚潮流/Sheet 2-Table 1.csv")
```

```python
dataset
```

```python
rmd = dataset[dataset["分桶"]=="个性化"]
```

```python
base = dataset[dataset["分桶"]=="非个性化"]
```

```python
from ggplot import *
import pylab as plt
import numpy
date = list(rmd["日期"])
click_ratio = [ float(x.strip("%"))/100 for x in list(rmd["点击UV转化率"])]
base_click_ratio = [ float(x.strip("%"))/100 for x in list(base["点击UV转化率"])]
print date, click_ratio, base_click_ratio
plt.plot(date, click_ratio, 'b-', label = u"personal")
plt.plot(date, base_click_ratio, 'r-', label = u"general")
plt.legend()
plt.show()
```
