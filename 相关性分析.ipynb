## 数据读取

```python

import pandas as pd
data_path = r"/Users/haoyuanhu/百度云同步盘/写作/inote/sz.csv"
df = pd.read_csv(data_path)
```

### 获取第二行的命名

```python
begin_name =  df.ix[0][1]
high_name = df.ix[0][2]
low_name = df.ix[0][3]
close_name = df.ix[0][4]
```

### 组成（股票名，价格类型）的命名

```python
names = list(df.columns)
length = len(names)/4
for i in range(length):
    name = names[(i*4 + 1)]
    names[i*4+1] = (name, begin_name)
    names[i*4+2] = (name, high_name)
    names[i*4+3] = (name, low_name)
    names[i*4+4] = (name, close_name)
```

### 更换命名， 删除多余数据

```python
df.columns = names
df.tail()
```

```python
df.drop(0)
```

### 计算每天的涨跌趋势

```python
ratio = {}
for k in df.columns:
    if k[1] != close_name:
        continue
    ratio[k[0]] =  list(df[(k[0], k[1])][1:])
```

```python
%doctest_mode
```

```python
import scipy.spatial as sl
from numpy import array
import numpy
table = {}
whole = {}
for k,v in ratio.items():
    rs = {}
    for k1, v1 in ratio.items():
        v = numpy.float32(v)
        v1 = numpy.float32(v1)
        a = array(v)
        b = array(v1)
        rs[k1] = sl.distance.cosine(a, b)
        whole[(k, k1)] = sl.distance.cosine(a, b)
    dict= sorted(rs.iteritems(), key=lambda d:d[1], reverse = True)
    table[k] = dict
whole = sorted(whole.iteritems(), key=lambda d:d[1], reverse = True)
```

```python
for k,v in table.items():
    if k.find("东源") >= 0:
        for v1 in v:
            print k,v1[0], v1[1]
```
