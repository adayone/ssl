# optimization

```python



def f(x):
    return x ** 2 + 10 * np.sin(x)

x = np.arange(-10, 10, 0.5)
#plt.plot(x, f(x))
#plt.show()
#plt.close()

# bfgs
o1 = optimize.fmin_bfgs(f, 0, disp=True)
print "global",  o1[0]
o2 = optimize.fmin_bfgs(f, 3, disp=True)
print "part", o2[0]

# grid
grid = (-10, 10, 0.1)
o3 = optimize.brute(f, (grid,))
print "grid", o3[0]

# grid part
o4 = optimize.fminbound(f, 0, 10)
print "grid part", o4

# root
root = optimize.fsolve(f, 1) 
print root
root = optimize.fsolve(f, -2.5) 
print root

# noise fit
xdata = np.linspace(-10, 10, num=100)
ydata = f(xdata) + (np.random.rand(xdata.size) - 0.5) * 10

def f2(x, a, b):
    return a * x ** 2 + b * np.sin(x)

guess = [-2, 2]
params, params_covariance = optimize.curve_fit(f2, xdata, ydata, guess)
print "params", params, "covar", params_covariance

# plot
plt.figure()
plt.plot(x, f(x), color='red', label='$f(x) = x^2 + 10 \cdot sin(x)$')
plt.show()
plt.plot(x, f2(x, params[0], params[1]), 'b--', label='fit $f(x) = x^2 + 10sin(x)$', linewidth=2)
plt.show()
#plt.plot(x, f2(x, 1, 1), 'g', label='$f(x) = x^2 + sin(x)$')
plt.plot(xdata, ydata, 'y', label='noise')
ox = np.array([o1[0], o2[0], o3[0], o4])
print ox
plt.plot(ox, f(ox), 'b^', label="min")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.show()

plt.figure()
plt.plot(x, f(x), color='red', label='$f(x) = x^2 + 10 \cdot sin(x)$')
plt.plot(x, f2(x, params[0], params[1]), 'b--', label='fit $f(x) = x^2 + 10sin(x)$', linewidth=2)
#plt.plot(x, f2(x, 1, 1), 'g', label='$f(x) = x^2 + sin(x)$')
plt.plot(xdata, ydata, 'y', label='noise')
ox = np.array([o1[0], o2[0], o3[0], o4])
print ox
plt.plot(ox, f(ox), 'b^', label="min")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.show()
```

# 代数计算

```python
from scipy import linalg
import numpy as np
# 计算方阵的行列式
arr = np.array([[0.1, 0, 0], [0.3, 0.4, 0], [0,0.5,0.2]])
print linalg.det(arr)
```

```python
# 计算方阵的逆
# 潜客数据挖掘
arr = np.array([[0.1, 0, 0], [0.3, 0.4, 0], [0,0.5,0.2]])

#print linalg.inv(arr)

tarr = np.transpose(arr)
print arr
print tarr
#print np.dot(tarr, arr)
s = np.dot(arr, tarr)
print np.dot(s, arr)
```

```python
# 计算奇异阵的逆, 会挂， 行列式为0
arr = np.array([[3,2], [6,4]])
print linalg.det(arr)
#print linalg.inv(arr)
```

```python
# SVD
arr = np.arange(9).reshape(3, 3) + np.diag([1,0,1])
print (linalg.svd(arr))
```

# 统计和随机数

```python
import numpy as np
from scipy.stats import norm
a = np.random.normal(size=1000)
```

```python
np.mean(a)
```

```python
np.average(a)
```

```python
bins = np.arange(-4, 5)
```

```python
histogram = np.histogram(a, bins=bins, normed=True)[0]
bins = 0.5*(bins[1:] + bins[:-1])
b = norm.pdf(bins)  # norm是正态分布
c = norm.cdf(bins)
```

```python
plt.figure()
plt.plot(bins, histogram, label="hisogram")
plt.plot(bins, b, label="pdf")
plt.plot(bins, c, label="cdf")
plt.legend()
plt.show()
```
