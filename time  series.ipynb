```python
import matplotlib.pyplot as plt
import datetime as dt
import pandas as pd
import pandas.io.data as web
import matplotlib
matplotlib.style.use('ggplot')
```

```python
for name in ('bzun', 'aapl', 'goog', 'bidu', 'baba', 'qihu'):
    baba = web.DataReader(name, "yahoo", dt.datetime(2014, 6, 1), retry_count=5, pause=0.1)
    baba['15_close'] = pd.stats.moments.rolling_mean(baba['Adj Close'], 15)
    baba['45_close'] = pd.stats.moments.rolling_mean(baba['Adj Close'], 45)
    fig = baba.plot(y=['Adj Close', '15_close', '45_close', 'Volume'], secondary_y='Volume', legend=True, figsize=(16, 10)).get_figure()
    fig.savefig('./image/%s.png'%name)
```
